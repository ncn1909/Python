```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://vnexpress.net/thoi-su'
response = requests.get(url)
#type(response.text)
soup = BeautifulSoup(response.text, 'html.parser')
#soup
```

```python
def get_article(article_div):
    title = article_div.find(class_ = "title-news")
    link = ''
    ti = ''
    if title is not None:
        ti = title.text.strip()
        link = title.a['href']
    des = ''
    description = article_div.find('p', 'description')
    if description != None:
        des = description.text.strip()
    dt = ''
    date = article_div.find('span', 'time-public')
    if date != None:
        dt = date.text.strip()

    return {'Title': ti, 'Description': des, 'Publish_Time': dt, 'Link': link}
```

```python
### Run your tests here
article_divs = soup.find_all('article')
data = [] 

print(len(article_divs))

for article_div in article_divs:
    row = get_article(article_div)
    data.append(row)
df = pd.DataFrame(data, columns = ['Title', 'Description', 'Publish_Time', 'Link'])
df.head(35)
```
